```r
#read data
data1=read.csv("cluster data.csv")
data=data1[,c(2,5)]
head(data)
```

```r
#elbow method to check optimal number of k
k.max <- 15 

wss <- sapply(1:k.max,function(k) {kmeans(data,
                                          k)$tot.withinss})
plot (1:k.max,wss,type = "b",frame = FALSE,
      xlab = "no of cluster k ",
      ylab = "total within cluster sum of square")
```

```r
#final model
k <- kmeans(data,3)
k
```

```r
#plot model
library(animation)
kmeans.ani(data,3)
```

```r
sum_ind = k$withinss  #sum of individual cluster points
sum_tot = k$tot.withinss   #sum of total cluster within summ
cent_points = k$centers   #center points of cluster
grps = k$cluster  # grroups of data points in cluster
```

```r
#write final data
finaldata<-cbind(data,k$cluster)
write.csv(finaldata,"cluster_output.csv")
```
